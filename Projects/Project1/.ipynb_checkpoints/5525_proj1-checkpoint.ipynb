7. For the customer churn prediction task, what is the best choice of k[1-10] in the k-nearest neighbor algorithm in the 10-fold cross-validation scenario?

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import sklearn.metrics as metrics
from sklearn.model_selection import cross_val_score

numNeighbors = range(1, 11) #test kNN classifcation by using a k range of 1-10
bestCrossScores = []

df = pd.read_csv("Experiment - 1/telecom_churn.csv")

# Data Pre-Processing
df['international plan'] = pd.factorize(df['international plan'])[0]
df['voice mail plan'] = pd.factorize(df['voice mail plan'])[0]
df['churn'] = df['churn'].astype('int')
df = df.drop(columns=["state", "phone number"])

# Assign Features vs Target variables
X = df.drop(columns=["churn"])
Y = df["churn"]


# Train a kNN model for each k value. Log the accuracy.
for k in numNeighbors:
    clf = KNeighborsClassifier(n_neighbors=k, metric='minkowski', p=2)
    scores = cross_val_score(clf, X, Y, cv=10)
    bestCrossScores.append(sum(scores)/len(scores))
```

```python
print("The best k (and its accuracy) for a 10-fold cross validation kNN model of the customer churn dataset is:" 
      "\n\tK = {}\n\tAccuracy = {:.4f}".format(bestCrossScores.index(max(bestCrossScores))+1, max(bestCrossScores)))
    
```

```python
plt.plot(numNeighbors, bestCrossScores, 'ro-')
plt.legend(['Training Accuracy','Test Accuracy'])
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
```

```python
print("Here is the list of bestCrossScores:")
bestCrossScores
```

5. Using bag of words approach, convert documents = ['Hi, how are you?', 'Win money, win from home. Call now.', 'Hi., Call you now or tomorrow?'] to its document-term matrix. 

```python
documents_orig = ['Hi, how are you?', 'Win money, win from home. Call now.', 'Hi., Call you now or tomorrow?']

# 1) Prep tokenizer by ...
#     setting all chars to lowercase and...
documents = [d.lower() for d in documents_orig]
```

```python
#     removing punctuation (IE: all ',' '?' '.' symbols).
import string

for index in range(len(documents)):
    documents[index] = documents[index].replace('.', '')
    documents[index] = documents[index].replace('?', '')
    documents[index] = documents[index].replace(',', '')
    
print(documents)
```

```python
# 2) Tokenize each word in each index of the sms list.
import numpy as np

list_of_words = []

for i in range(len(documents)):
    list_of_words.append(documents[i].split())

print(list_of_words)
```

6. Report accuracy, precision, recall and F1 score for the spam class after applying Naive Bayes algorithm.

** Solution to 3.6 **

Below is the resulting accuracy, precision, recall and F1 score of a multinomial Naive Bayes model's ability to classify real or SPAM phone SMS messages.

Accuracy score: 0.98325
Precision score: 0.95673
Recall score: 0.91284
F1 score: 0.93427

```python
# 1) Create Training:Test data

from sklearn.model_selection import train_test_split

df['label'] = df['label'].replace('ham', 0)
df['label'] = df['label'].replace('spam', 1)

X_train, X_test, y_train, y_test = train_test_split(df['sms'], 
                                                    df['label'],test_size=0.30, 
                                                    random_state=1)
```

```python
# 2) Instantiate the CountVectorizer method to create Frequence_Matrix tables for Training & Test data
from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer()

# Fit the training data and then return the matrix
training_data = count_vector.fit_transform(X_train)

# Transform testing data and return the matrix. 
testing_data = count_vector.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes = MultinomialNB()
naive_bayes.fit(training_data, y_train)
```

```python
# 3) Make prediction on test dataset
predictions = naive_bayes.predict(testing_data)
```

```python
# 4) Determine the performance of the multinomial Naive Bayes model
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('Accuracy score: {:.5f}'.format(accuracy_score(y_test, predictions)))
print('Precision score: {:.5f}'.format(precision_score(y_test, predictions)))
print('Recall score: {:.5f}'.format(recall_score(y_test, predictions)))
print('F1 score: {:.5f}'.format(f1_score(y_test, predictions)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions, normalize='all')
```
